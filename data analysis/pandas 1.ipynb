## Missing values

```python
NewDF = {'A': [1,2, np.nan], 'B':[5, np.nan, np.nan], 'C': [1,2,3]}
```

```python
newDF = pd.DataFrame(NewDF)
newDF
```

```python
newDF.isnull()
```

```python
newDF.dropna(axis= 1)
```
