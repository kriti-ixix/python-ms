```python
import matplotlib.pyplot as plt
import numpy as np


%matplotlib inline
```

## Line plot

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, 'g', linestyle = 'dashed', linewidth = 2, marker = '*', markersize = 13)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D graph')
```

```python
x = np.linspace(1,5,11)
y =  x**2
```

```python
x
```

```python
plt.plot(x, y, color = 'r',marker = 'o', linewidth = 1)
```

```python
plt.plot(x,y, 'g*--')  # here you can summerize color,marker and linestyle parameter
```

## Bar plot

```python
x = [2,8,10]
y = [11,16,9]
```

```python
x2 = [3,9,11]
y2 = [6,15,7]
```

```python
plt.bar(x,y)
plt.bar(x2, y2)
plt.title('Bar plot', color = 'g')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

```python
plt.bar(x,y)
plt.bar(x2, y2)
plt.title('Bar plot', color = 'g')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

## Hist plot

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('histogram')
plt.show()
```

## Box Plot

```python
data = [np.random.normal(0,std, 100) for std in range(1,4)]
plt.boxplot(data)
plt.show()
```

```python
data
```

## Pie plot

```python
plt.figure(figsize = (5,6))
sizes = [215, 130, 245, 210]
labels = ['python', 'C++', 'Ruby','Java']
explode = [0.2, 0.4,0,0]
plt.pie(sizes, labels = labels, explode = explode , autopct = '%1.2f%%')
plt.show()
```
